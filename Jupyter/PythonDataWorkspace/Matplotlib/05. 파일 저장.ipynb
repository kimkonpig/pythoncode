# 5. 파일 저장

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.savefig('graph.png', dpi=100)
```

```python
plt.figure(dpi=200)
plt.plot(x, y)
plt.savefig('graph_200.png', dpi=100)
```
