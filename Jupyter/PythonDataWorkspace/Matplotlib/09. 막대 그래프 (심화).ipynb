# 9. 막대 그래프 (심화)

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # //// 모양으로 채우기
bar[1].set_hatch('x') # xxxx 모양으로 채우기
bar[2].set_hatch('/*') # // + ** 모양으로 채우기
# 구글에 matplotlib hatch 검색하면 document 확인 가능
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='b')
```
