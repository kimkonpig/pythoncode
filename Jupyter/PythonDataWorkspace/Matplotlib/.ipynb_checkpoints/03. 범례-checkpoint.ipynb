# 3. 범례(legend)

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='best') # 최적의 위치
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc=(0.7, 0.8)) # x축, y축 (0~1 사이값)
```
