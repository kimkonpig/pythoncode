# 4. 스타일

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5) # 선 두께 설정
```

## 마커(marker)

```python
plt.plot(x, y, marker='o') # 마커 추가
```

```python
plt.plot(x, y, marker='o', linestyle='None') # 마커만 남고 선은 없음
```

```python
plt.plot(x, y, marker='v') # 삼각형 마커
```

```python
plt.plot(x, y, marker='v', markersize=10) # 삼각형 마커 사이즈 변경
```

```python
plt.plot(x, y, marker='X', markersize=10) # X표 마커
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red') # 마커 외곽선 색상
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow') # 마커 내부 색상 변경
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':') # 자잘한 점선
```

```python
plt.plot(x, y, linestyle='--') # 길쭉한 점선
```

```python
plt.plot(x, y, linestyle='-.') # -.-. 형태의 선
```

```python
plt.plot(x, y, linestyle='-') # 실선
```

## 색깔

```python
plt.plot(x, y, color='pink') # 선 색깔 핑크
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b') # matplotlib에서 정의해둔 색상이름(b, g 등...)
```

```python
plt.plot(x, y, color='g')
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color=빨간색, marker=o, linestyle=점선
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='none')
```

## 축약어
구글에 matplotlib markerfacecolor mfc 로 검색하면 document 확인 가능

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='black', ls=':') # markerfacecolor, markersize, markeredghcolor, linestyle
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.4) # 투명도(0~1 사이값)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=200) # dots per inch
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='pink')
plt.plot(x, y)
```
