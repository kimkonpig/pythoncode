# 6. 텍스트

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, marker='o')

for idx, txt in enumerate(y): # y데이터 [2, 4, 8]
    plt.text(x[idx], y[idx] + 0.3, txt, ha='center', color='blue') # (x, y)에 텍스트 표시
```
